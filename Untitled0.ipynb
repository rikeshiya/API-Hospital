```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
data = {
    "Area (sqft)": [450,480,500,420,480,500,470,450,460,480,500,520,470,450,430,440,400,420,410,430,480,500,520,550,480,470,500,520,450,480],
    "Price per Sqft (INR)": [55000,60000,52000,58000,65000,62000,60000,50000,48000,52000,45000,44000,55000,47000,50000,48000,45000,43000,42000,41000,46000,48000,45000,50000,44000,43000,58000,54000,42000,56000],
    "Property Price (INR Crore)": [2.48,2.88,2.6,2.44,3.12,3.1,2.82,2.25,2.21,2.5,2.25,2.29,2.59,2.12,2.15,2.11,1.8,1.81,1.72,1.76,2.21,2.4,2.34,2.75,2.11,2.02,2.9,2.81,1.89,2.69]
}

df = pd.DataFrame(data)

# Calculate central tendencies
mean_values = df.mean()
median_values = df.median()
mode_values = df.mode().iloc[0]

# Print results
print("Mean:\n", mean_values)
print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)


# Function to plot central tendency graph
def plot_central_tendency(column_name):
    plt.figure()
    values = [
        mean_values[column_name],
        median_values[column_name],
        mode_values[column_name]
    ]

    plt.bar(["Mean", "Median", "Mode"], values)
    plt.title(f"Central Tendency - {column_name}")
    plt.ylabel(column_name)
    plt.show()


# Plot graphs separately
plot_central_tendency("Area (sqft)")
plot_central_tendency("Price per Sqft (INR)")
plot_central_tendency("Property Price (INR Crore)")
```
